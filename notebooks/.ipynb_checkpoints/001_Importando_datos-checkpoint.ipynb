```python
import pandas as pd
from iertools.read import read_epw
```

```python
f = '../data/MEX_MOR_Cuernavaca-Matamoros.Intl.AP.767260_TMYx.2004-2018.epw'
data = read_epw(f,alias=True,year=2023)
data.columns
```

```python
columnas = ['To','RH', 'P',  'Ig', 'Ib', 'Id', 'Wd', 'Ws']
tmx  = data[columnas]
tmx.plot(subplots=True,figsize=(12,10))
```

```python
tmx.to_csv('../data/tmx.csv')
```
