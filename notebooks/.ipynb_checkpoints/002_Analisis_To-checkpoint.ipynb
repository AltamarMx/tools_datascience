```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = '../data/tmx.csv'
data = pd.read_csv(f)
# Convert 'tiempo' to datetime format
data['tiempo'] = pd.to_datetime(data['tiempo'])
# Extract hour of day
data['hora_del_dia'] = data['tiempo'].dt.hour
```

## All days in one day

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.scatter(data['hora_del_dia'], data['To'], alpha=0.2)

ax.set_xlabel('Hora del da')
ax.set_ylabel('To')
ax.grid()
```

```python
data['fecha'] = data['tiempo'].dt.date
dates = data['fecha'].unique()
# Identify the day with the maximum 'To' value
max_to_date = data[data['To'] == data['To'].max()]['fecha'].values[0]

# Identify the day with the minimum 'To' value
min_to_date = data[data['To'] == data['To'].min()]['fecha'].values[0]


# Filter the data to include only the times between 4 and 8 a.m.
morning_data = data[(data['hora_del_dia'] >= 4) & (data['hora_del_dia'] <= 8)]

# Identify the day with the minimum 'To' value in this time range
min_to_date_morning = morning_data[morning_data['To'] == morning_data['To'].min()]['fecha'].values[0]

# Set figure size
plt.figure(figsize=(10, 6))

# Plot 'To' against 'hora_del_dia' for each date
for date in dates:
    data_day = data[data['fecha'] == date]
    if date == max_to_date:
        plt.plot(data_day['hora_del_dia'], data_day['To'], color='red', label='Mxima To', alpha=0.6)
    elif date == min_to_date_morning:
        plt.plot(data_day['hora_del_dia'], data_day['To'], color='blue', label='Mnima To (4-8 a.m.)', alpha=0.6)
    else:
        plt.plot(data_day['hora_del_dia'], data_day['To'], color='grey', alpha=0.1)

plt.xlabel('Hora del da')
plt.ylabel('To')
plt.title('Valores de To por hora del da')
plt.legend()

# Set x-axis limits and ticks
plt.xlim(0, 24)
plt.xticks(range(0, 25, 6))

plt.grid(True)
plt.show()
```
